```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df =pd.read_csv("clean_bike_data.csv")
df.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
import sklearn
```

```python
#df.column
x=df.drop('price',axis=1)
y=df[['price']]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

## LinearRegression

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
pred = lr.predict(x_test)
```

```python
pred[0]
```

```python
y_test
```

## svm

```python
from sklearn.svm import SVC,SVR
sv_bike=SVC()
```

```python
sv_bike.fit(x_train,y_train)
```

```python
pred=sv_bike.predict(x_test)
```

```python
pred
```

```python
print("accuracy of test data :-",sv_bike.score(x_test,y_test)*100)
print("accuracy of train data :-",sv_bike.score(x_train,y_train)*100)
```

```python
y_test
```

# Decision Tree (DT)

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt=DecisionTreeClassifier()
dt.get_params()
```

```python
dt.fit(x_train,y_train)
```

```python
print("training accuracy :-", dt.score(x_train,y_train))
print("testing accuracy :-", dt.score(x_test,y_test))
```

```python
# with parameter
from sklearn.tree import DecisionTreeClassifier
dt_with_para =DecisionTreeClassifier(criterion='log_loss',min_samples_split=3,min_samples_leaf=2)
```

```python
dt_with_para.fit(x_train,y_train)
```

```python
print("training accuracy :-",dt_with_para.score(x_train,y_train))
print("testing accuracy :-",dt_with_para.score(x_test,y_test))
```

## Save ML model

```python
import joblib
joblib.dump(dt_with_para,"dt_bike_sytem_model.lb")
```

```python
model=joblib.load("dt_bike_sytem_model.lb")
```
